Elements: HTML elements are the building blocks of a web page. They define the structure and content of the page. An HTML element usually consists of a start tag, content, and an end tag. For example, <p> is the start tag for a paragraph element, and </p> is the end tag.

Tags: HTML tags are used to define elements in a markup language. They consist of angle brackets enclosing the tag name. Tags come in pairs: an opening tag and a closing tag. The opening tag precedes the content, and the closing tag follows the content. For example, <p> is an opening tag, and </p> is a closing tag.

Class: The class attribute is used to assign one or more class names to an HTML element. Classes are used in CSS to style and manipulate elements with the same class name. Multiple elements can share the same class, allowing you to apply the same styles to all of them. For example, <div class="container"> assigns the class "container" to a <div> element.

Attributes: HTML attributes provide additional information about an element and are specified within the start tag of an element. Attributes consist of a name and a value, separated by an equals sign (=). They modify the behavior or appearance of the element. Common attributes include id, class, href, src, alt, etc. For example, in <img src="image.jpg" alt="Description">, src and alt are attributes.

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
#since the data I collected doesn't have the hotel details I need. I tried to webscrape from travel sites. 
# I used the traveloka website to get latest data of hotels in Puerto princesa. Traveloka is the largest online travel app in Southeast Asia.
url = "https://www.traveloka.com/en-ph/hotel/philippines/city/puerto-princesa-10011051"

# Send an HTTP GET request to the URL.
response = requests.get(url)

# Check if the request was successful (status code 200).
if response.status_code == 200:
    # Print the HTML content of the page.
    print(response.text)
else:
    print(f"Failed to retrieve the page. Status code: {response.status_code}")
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
print(soup.prettify())
```

```python
# Get all tags 
tags = {tag.name for tag in soup.find_all()} 

# class list set 
class_list = set() 
  
# iterate all tags 
for tag in tags: 
  
    # find all element of tag 
    for i in soup.find_all( tag ): 
  
        # if tag has attribute of class 
        if i.has_attr( "class" ): 
  
            if len( i['class'] ) != 0: 
                class_list.add(" ".join( i['class'])) 
  
print(tags)
print( class_list ) 
```

```python
# Find all hotel listings
hotel_listings = soup.find_all('div', class_='css-1dbjc4n r-6koalj r-13awgt0 r-1cmwbt1 r-1ik5qf4 r-nsbfu8')
#css-1dbjc4n
print(hotel_listings)
```

```python
for listing in hotel_listings:
    hotel_name_elements = listing.find_all('h3')

print(hotel_name_elements)
```

```python
# Addtl info. Extract the multiple attribute values

# Get the whole h3 tag 
tag = soup.h3 
  
# Get the attribute 
attribute = tag['class'] 
  
# Print the output 
print(attribute)
```

```python
hotel_names = []
   
# Extract and append the hotel names to the hotel_names list
for element in hotel_name_elements:
    hotel_names.append(element.get_text())

print(hotel_names)
```

```python
hotel_types = []
divs = soup.find_all("div", {"class": "css-1dbjc4n r-yonkja"})
for li in divs:
    hotel_types.append(li.getText())
print(hotel_types)
```

"div": This is the name of the HTML tag you're searching for. In this case, you're looking for <div> elements.

{"class": "css-1dbjc4n r-yonkja"}: This is a dictionary specifying the attributes you're looking for within the <div> elements. Here, you're looking for <div> elements that have both the classes "css-1dbjc4n" and "r-yonkja". In HTML, classes are used to apply styles or group elements together, and elements can have multiple classes separated by spaces.

```python
hotel_rating = []
# Find all <meta> elements with itemprop="ratingValue"
meta_elements = soup.find_all('meta', itemprop='ratingValue')

# Extract the values of the content attribute for these <meta> elements
ratings = [element['content'] for element in meta_elements]

for rating in ratings:
    hotel_rating.append(rating)
#Last item have no rating
hotel_rating.append('0')
print(hotel_rating)
```

```python
hotel_address = []
divs = soup.find_all("div", {"class": "css-901oao css-bfa6kz r-t1w4ow r-1b43r93 r-majxgm r-rjixqe r-fdjqy7"})
for li in divs:
    hotel_address.append(li.getText())
print(hotel_address)
```

```python
hotel_rate = []
divs = soup.find_all("div", {"class": "css-901oao r-1i6uqv8 r-t1w4ow r-1b43r93 r-majxgm r-rjixqe r-fdjqy7"})
for li in divs:
    hotel_rate.append(li.getText())
print(hotel_rate)
```

```python
hotel_review_count = []
divs = soup.find_all("div", {"class": "css-901oao r-t1w4ow r-1b43r93 r-majxgm r-rjixqe r-fdjqy7"})
for li in divs:
    hotel_review_count.append(li.getText())
print(hotel_review_count)
```

```python
hotel_topreview = []
divs = soup.find_all("div", {"class": "css-1dbjc4n r-1udh08x"})
for li in divs:
    hotel_topreview.append(li.getText())
hotel_topreview.append('Blank')
print(hotel_topreview)
```

```python
# Create a DataFrame
data = {
    'Hotel Name': hotel_names,
    'Type': hotel_types,
    'Rating': hotel_rating,
    'Address': hotel_address,
    'Top Review': hotel_topreview
}
hotels = pd.DataFrame(data)

# Display the DataFrame
hotels
```
