Used requests library in making HTTP requests to a specified URL and get the HTML code of the website. BeautifulSoup module was used to parse the html and then make it pretty, then used the find_all(), get_text() to scrape.

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url =  "https://wanderlog.com/list/geoCategory/70567/best-places-to-eat-in-puerto-princesa"

# Send an HTTP GET request to the URL.
response = requests.get(url)

# Check if the request was successful (status code 200).
if response.status_code == 200:
    # Print the HTML content of the page.
    print(response.text)
else:
    print(f"Failed to retrieve the page. Status code: {response.status_code}")
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
print(soup.prettify())
```

```python
resto_listings = soup.find_all('div', class_='mt-2')
print(resto_listings)
```

```python
resto_names = []
for listing in resto_listings:
    resto_name_elements = listing.find_all('h2')
print(listing)
```

```python
resto_names = []
for listing in resto_listings:
    resto_name_elements = listing.find_all('h2')
    
    # Extract and append the hotel names to the hotel_names list
    for element in resto_name_elements:
        resto_names.append(element.get_text())

# Now, the hotel_names list should contain all the extracted hotel names
print(resto_names)
```

```python
star_rating = []
for listing in resto_listings:
    star_rating_elements = listing.find_all("span", {"class": "font-weight-bold"})
    
    # Extract and append the hotel names to the hotel_names list
    for element in star_rating_elements:
        star_rating.append(element.get_text())


# The decimal_numbers list will contain only the decimal numbers from the original_list
star_rating = [x for x in star_rating if x.replace('.', '', 1).isdigit()]
star_ratingT= star_rating[::2]  # Get even-indexed elements
star_ratingG = star_rating[1::2]  # Get odd-indexed elements

print("Even-indexed elements:", star_ratingT)
print("Odd-indexed elements:", star_ratingG)
```

```python
rating_count = []
for listing in resto_listings:
    rating_count_elements = listing.find_all("span", {"class": "ml-1 text-muted"})
    
    # Extract and append the hotel names to the hotel_names list
    for element in rating_count_elements:
        text = element.get_text().replace("(", "").replace(")", "")
        rating_count.append(text)

# Now, the hotel_names list should contain all the extracted hotel names
rating_countT= rating_count[::2]  # Get even-indexed elements
rating_countG = rating_count[1::2]  # Get odd-indexed elements

print("Even-indexed elements:", rating_countT)
print("Odd-indexed elements:", rating_countG)
```

```python
pricing = []

for listing in resto_listings:
    resto_listing2 = listing.find("div", {"class":"badge Badge__light text-nowrap Badge__shape-pill mr-1"})
    
    if resto_listing2:
        # Within 'resto_listing', find all elements with class "text-muted"
        pricing_elements = resto_listing2.find_all("span", {"class": "text-muted"})

        # Extract and append the text of each 'rating_countG_element' to the 'rating_countG' list
        for element in pricing_elements:
            pricing.append(element.get_text())

# Now, the 'rating_countG' list should contain all the extracted text from the first matching element in each 'resto_listings' element
print(len(pricing))
```

```python
pricing = []

for listing in resto_listings:
    resto_listing2 = listing.find("div", {"class":"badge Badge__light text-nowrap Badge__shape-pill mr-1"})
    
    if resto_listing2:
        # Within 'resto_listing', find all elements with class "text-muted"
        pricing_elements = resto_listing2.find_all("span", {"class": "text-muted"})

        # Extract and append the text of each 'rating_countG_element' to the 'rating_countG' list
        for element in pricing_elements:
            if element:
                pricing.append(element.get_text())
            else:
                pricing.append(None)

# Now, the 'rating_countG' list should contain all the extracted text from the first matching element in each 'resto_listings' element
print(len(pricing))
```
